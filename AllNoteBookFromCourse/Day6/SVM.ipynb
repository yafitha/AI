## Gaining Intuition

<p style="line-height:1.75;font-size:16px">
We'll start with a basic example where we have a single feature and we that we want to use in order to separate (classify) our data. Let's go back to our sharks and dolphins data and try to classify our sample using only the length. We'll put all of our data points on an axis (dolphin=blue, shark=red) and choose a threshold which will separate our data well (marked with a yellow line)

<p style="line-height:1.75;font-size:16px">
According to our training data the model is excellent giving us 100% accuracy and now anytime a new observation comes in we can simply compare it to our threshold - if it's lower than the threshold it's a dolphin, if it's higher than the threshold it's a shark. However, what happens if a the following observation comes in?

<p style="line-height:1.75;font-size:16px">
Based on our threshold, it should be classified as a shark, however, it seems to be closer to the dolphin observations than it is to the sharks. Perhaps our threshold was incorrect to begin with even though it gave us good results. Let's look at the two closest observations from different classes and try to set the threshold directly between them.

<p style="line-height:1.75;font-size:16px">
Now, any new observation that comes in an falls to the left of the threshold will be closer to the other dolphin observations and classified as a dolphin. Similarly any new observation that comes in an falls to the right of the threshold will be closer to the other shark observations and classified as a shark.<br>
The shortest distance between the observations and the threshold is called the <b>margin</b>. Notice that since the threshold is placed in the middle of the observations, their distance to the threshold is identical and the margin is as large as it can be. Any movement of the threshold, left or right, will make the margin smaller. When we use this type of margin for classification we called is a <b>Maximal Margin Classifier</b>.<br> Maximal margin classifiers seem to be the best we can do, but consider what would have happened if our training looked like this?

<p style="line-height:1.75;font-size:16px">
We would have to change our threshold so that it will be in between the two closest observations from different classes.

<p style="line-height:1.75;font-size:16px">
Now, the maximal margin classifier is closer to the shark observations than it is to the dolphin observations and that leads to another issue. Consider the following new observation coming in.

<p style="line-height:1.75;font-size:16px">
It seems to be closer to a shark than it is to a dolphin however, according to the maximal margin classifier, this observation would be classified as a dolphin. The takeaway from this is that maximal margin classifiers are very sensitive to outliers in the data.

In the case where we allow for misclassifications, the distance between the two closest observations from different classes (without the misclassifications) is called a <b>soft margin</b>.
Let's pause for a second to write down some of our findings. When we used the maximal margin classifier, we had a classifier that was very sensitive to the training data (low bias) and had bad performance on new data (high variance). When we allowed for some misclassifications we got a classifier that was less sensitive to the training data (higher bias) and performed better on new data (lower variance).<br>
When we use a soft margin to determine the threshold, we are using a <b>Support Vector Classifier</b> and the observations that helped us set the margin are called the <b>Support Vectors</b>

But how do we find the correct transformation? We use a <b>kernel function</b> to systematically find support vector classifiers in higher dimensions. Two widely used kernels are the <b>polynomial kernel</b> and the <b>radial kernel</b>.

### Polynomial Kernel

<p style="line-height:1.75;font-size:16px">
The polynomial kernel take the form of:
<center><h3>
$(a\times b+r)^d$
</h3>
</center>
<p style="line-height:1.75;font-size:16px">
Where $a$ and $b$ are two different observations in the dataset, $r$ is the coefficient of the polynomial and $d$ is the degree of the polynomial. Let's set $r=\frac{1}{2}$ and $d=2$ and look at an example to understand this better.
<center><h3>
$(a\times b+\frac{1}{2})^2=(a\times b+\frac{1}{2})(a\times b+\frac{1}{2})=ab+a^2b^2+\frac{1}{4}=(a, a^2, \frac{1}{2})(b, b^2, \frac{1}{2})$
</h3>
</center>
<p style="line-height:1.75;font-size:16px">
We can look at this output a coordinates for (x, y, z) in our new vector space (we'll ignore z for now since it's identical). What this mean is that we can take something that looks like this:

### Radial Kernel (RBF)

<p style="line-height:1.75;font-size:16px">
The radial basis function finds support vector classifiers in infinite dimensions. It takes the form of:
<center><h3>
$e^{-\gamma(a-b)^2}$
</h3>
</center>
<p style="line-height:1.75;font-size:16px">
Since it works in infinite dimensions we won't be able to visualize it but we'll try to get an idea of what it does. Without going too much into the details, the radial kernel behaves like a weighted nearest neighbor algorithm when examining a new observation. Observations that are close to the new observation will have more influence on how we classify it in comparison to observations that are further away. The influence that each observation contributes to the classification is determined first by $(a-b)^2$ the squared distance between the two observations. Second, $\gamma$ scales the influence further - smaller $\gamma$ means more influence.

## Pros and Cons

<p style="line-height:1.75;font-size:16px">
<b>Pros</b><br>
1\. Works really well with a clear margin of separation.<br>
2\. Effective in high dimensional spaces.<br>
3\. Effective in cases where the number of dimensions is greater than the number of samples.<br>
4\. Memory efficient.<br>

<p style="line-height:1.75;font-size:16px">
<b>Cons</b><br>
1\. Performance becomes an issue on large datasets.<br>
2\. Does not perform well when data is noisy and targets are overlapping.<br>

## SVM with SKLearn

### Example 1

<p style="line-height:1.75;font-size:16px">
We'll generate synthetic data so we can see SVM in action

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline

X, y = make_blobs(n_samples=50, centers=2, random_state=0, cluster_std=0.60)
plt.figure(figsize=(20,10))
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');
```

<p style="line-height:1.75;font-size:16px">
Now let's plot some possible SVC decision boundaries (we haven't trained anything yet, these are merely possible options)

```python
import numpy as np

xfit = np.linspace(-1, 3.5)

plt.figure(figsize=(20,10))
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')

for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',
                     color='#AAAAAA', alpha=0.4)

plt.xlim(-1, 3.5);
```

```python
from sklearn.svm import SVC

model = SVC(kernel='linear', C=100000)
model.fit(X, y)
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
plt.figure(figsize=(20,10))
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plot_svc_decision_function(model);
```

<p style="line-height:1.75;font-size:16px">

<p style="line-height:1.75;font-size:16px">
Let's see how the number of observations and the C value affect the model

```python
from ipywidgets import interact, fixed

def plot_svm(N=10, C=1E10, ax=None):
    X, y = make_blobs(n_samples=1000, centers=2,
                      random_state=0, cluster_std=0.60)
    X = X[:N]
    y = y[:N]
    model = SVC(kernel='linear', C=C)
    model.fit(X, y)
    
    fig = plt.figure(figsize=(20,10))
    ax = ax or plt.gca()
    ax.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
    ax.set_xlim(-1, 4)
    ax.set_ylim(-1, 6)
    plot_svc_decision_function(model, ax)

interact(plot_svm, N=[10, 50, 100, 200, 500, 1000], C=[0.1, 1, 10, 100, 1000, 10000, 100000], ax=fixed(None));
```
