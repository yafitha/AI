# K-Nearest Neighbor (KNN)

## Introduction

<p style="line-height:1.75;font-size:16px">
KNN is a non-linear supervised classification and regression method invented in 1951. Using the notion that "birds of a feather flock together", KNN relies on the fact that similar objects are likely to be found in close proximity to one another. In the classification case, when a new sample is given as input, KNN looks the closest K neighbors of that sample (using a distance metric) and classifies it according to the majority class of its nearest neighbors. Let's try to get an intuition of how this works by looking at an example.

In this example we had features with different scales and even different units. Although it didn't greatly affect our results in this case, it could give us incorrect results in other cases. This had to do with the way we calculate the distance between samples. If one feature in on a scale of 0-1 and another is on a scale of 1000-1500, a 500% difference in the first feature won't affect the distance metric as much as a 0.1% change in the second feature. That's why it's import to normalize/standardize our features when using algorithms that are affected by feature scale.

## KNN Algorithm

<p style="line-height:1.75;font-size:16px">
1\. Initialize K to your chosen number of neighbors<br>
2\. For each example in the data<br>
&nbsp;&nbsp;&nbsp;    2\.1\. Calculate the distance between the query example and the current example from the data<br>
&nbsp;&nbsp;&nbsp;    2\.2\. Add the distance and the index of the example to an ordered collection<br>
3\. Sort the ordered collection of distances and indices from smallest to largest (in ascending order) by the distances<br>
4\. Pick the first K entries from the sorted collection<br>
5\. Get the labels of the selected K entries<br>
6\. Return the most commonly observed value (mode) of the K labels

## Distance Metrics

<p style="line-height:1.75;font-size:16px">
KNN supports various metrics each with their own advantages and limitations. The most commonly used distance metric by far in the euclidean distance - it's easy to understand and calculate and tends to give good results. <a href=https://homes.cs.washington.edu/~pedrod/papers/cacm12.pdf>Some papers</a> suggest that Manhattan distance work better when we're dealing with high-dimenstional data.

<p style="line-height:1.75;font-size:16px">
Choosing the correct k value in KNN is important since it can drastically change our results. Unfortunately, there are no pre-defined statistical methods that help us choose this value but there are some guidelines:<br>
1\. A small k value leads to noisy decision boundries.<br>
2\. Large k values will lead to long computation times.<br>
3\. Using $\sqrt{n}$ where $n$ is the number of data points is a good starting point.<br>
4\. Using an odd k is better than an even one.<br>
<p style="line-height:1.75;font-size:16px">
Another option that is sometimes used is to run the algorithm with different values of k and record the error rate for each of this values. Then, simply choose a k which on one hand gives a low error rate and on the other hand isn't too large that it hurts performance.

## Pros and Cons

<p style="line-height:1.75;font-size:16px">
<b>Pros</b><br>
1\. <b>Pretty intuitive and simple</b> - KNN algorithm is very simple to understand and equally easy to implement.<br> 
2\. <b>It constantly evolves</b> - The classifier immediately adapts as we collect new training data. It allows the algorithm to respond quickly to changes in the input during real-time use.<br> 
3\. <b>Can be used both for classification and regression</b><br> 
4\. <b>One Hyper Parameter</b> - We only have to choose a value for k which save us time in hyperparameter tuning.<br> 

<p style="line-height:1.75;font-size:16px">
<b>Cons</b><br>
1\. <b>KNN is slow</b> - The algorithm might be very easy to implement but as dataset grows efficiency or speed of it declines very fast.<br>
2\. <b>Curse of Dimensionality</b> - KNN works well with small number of input variables but as the number of variables grow the algorithm struggles to predict the output of new data point.<br>
3\.<b>Requires homogeneous features</b> -  It is necessary that features have the same scale.<br>
4\.<b>Imbalanced data causes problems</b> - KNN doesn’t perform well on imbalanced data. The model will give a lot of preference to the majority class.<br>
5\.<b>Outlier sensitivity</b> - KNN is very sensitive to outliers as it simply chooses the neighbors based on distance criteria.


## KNN in SKLearn

```python
import pandas as pd 

data = pd.read_csv('./data/weight-height.csv')

inch_to_cm = 2.54
pound_to_kg = 0.453592

data['Height'] *= inch_to_cm
data['Weight'] *= pound_to_kg

data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(data.drop('Gender', axis=1), data['Gender'])

k = int(np.sqrt(len(X_train)))
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)

errors = []
for k in range(1,200):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    errors.append(1 - knn.score(X_test, y_test))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(20,10))
ax = fig.gca()
ax.set_ylabel('Error Rate')
ax.set_xlabel('k')

ax.plot(errors)
plt.show()
```
