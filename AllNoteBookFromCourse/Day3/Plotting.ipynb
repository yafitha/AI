### Figures and axes

<p style="line-height:1.75;font-size:16px">
There are three main components of a plot that you should know about:<br>
1.<span style="line-height:1.75;font-size:16px">
 <b>Figure</b> - the overall window or page that everything is drawn on.<br>
2.<span style="line-height:1.75;font-size:16px"> <b>Axes</b> - the area on which the data is plotted - what you would consider a plot. A Figure may have multiple Axes but an Axes can only belong to one figure.<br>
3.<span style="line-height:1.75;font-size:16px"> <b>Axis</b> - the number-line-like objects in the plot. An Axes contains two (or three in the 3D case) Axis objects.<br>

## Multiple Plots

<p style="line-height:1.75;font-size:16px">
Sometimes we want to show two or more lines in the same plot. Let's try to do that using what we've learned.

```python
x = np.linspace(-10, 10, 100)
y1 = x**2
y2 = x**3

fig = plt.figure(figsize=(15, 8))

ax = fig.gca()
ax.set_title('Multiple Functions')
ax.set_ylabel('y1')
ax.set_xlabel('x')

# We use twinx() to create another, independent y-axis
ax2 = ax.twinx()
ax2.set_ylabel('y2')

# Here we use b- to set the marker color to blue and also the marker shape to be a line (-)
line1, = ax.plot(x, y1, 'b-')
line2, = ax2.plot(x, y2, 'r.')
ax2.legend([line1, line2], ['X^2', 'X^3'])

# set_xlim limits the range of our x-axis
ax.set_xlim((-12,12))

plt.show()
```

<p style="line-height:1.75;font-size:16px">
Since we'll be using Numpy a alot, it is also useful to be able to plot data directly from a Numpy array. Luckily the `plot` method supports the use of numpy arrays.

```python
x = np.linspace(-3, 3, 100)
data = np.array([x, x**2, x**3]).T

fig = plt.figure(figsize=(15, 8))

ax = fig.gca()

ax.set_title('Functions Using Numpy')
ax.set_ylabel('y')
ax.set_xlabel('x')

line1, line2 = ax.plot(data[:, 0], data[:, 1:])

line1.set_linewidth(5)
line1.set_linestyle('-')
line1.set_color('b')

line2.set_linestyle(' ')
line2.set_marker('.')
line2.set_markersize(8)
line2.set_color('r')

ax.legend(['X^2', 'X^3'])

plt.show()
```

## Subplots

<p style="line-height:1.75;font-size:16px">
Subplots allow us to plot multiple Axes in the same Figure. Using Matplotlib's comprehensive api, we create the subplots and arrange them in any way we'd like.

```python
x = np.linspace(-3, 3, 100)
y1 = x**2
y2 = x**3

fig = plt.figure(figsize=(15,8))

# We use add_subplot with parameters: number of rows, number of columns, index of plot
ax1 = fig.add_subplot(2, 1, 1)
ax1.set_title('Subplot 1')
ax1.set_ylabel('y')
ax1.set_xlabel('x')

ax2 = fig.add_subplot(2, 1, 2)
ax2.set_title('Subplot 2')
ax2.set_ylabel('y')
ax2.set_xlabel('x')

line1, = ax1.plot(x, y1, 'b--')
line2, = ax2.plot(x, y2, 'r.')

line1.set_linewidth(10)
line2.set_markersize(7)

ax1.legend(['X^2'])
ax2.legend(['X^3'])

# Adjusts the padding of plots so that they fit nicely in the figure
fig.tight_layout()

plt.show()
```

## Another way to plot

<p style="line-height:1.75;font-size:16px">
Up until now we've onlly used `plt` as a way to access components such as `Figure` and `Axes` and then create the plot using those. However, `plt` offers a substantial api that lets you create plots directly without having to use the subcomponents. Using `plt` the various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current axes.

```python
x = np.linspace(-3, 3, 100)
y = x**2
# Note the figure size has to be defined before plotting the data
plt.figure(figsize=(15,8))
plt.plot(x, y, 'b.', markersize=10)
plt.title('A Function')
plt.ylabel('y')
plt.xlabel('x')
plt.legend(['X^2'])
plt.show()
```

# Plot Types

<p style="line-height:1.75;font-size:16px">
Up until now we've seen the most common type of plot which lets us plot y versus x as lines. Now, we'll look at some of the other types of plot that `matplotlib` supports.

## Scatter plot

<p style="line-height:1.75;font-size:16px">
A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables. `matplotlib` offers us the `scatter` method to easily create scatter plots.

```python
import pandas as pd

df = pd.read_csv('./data/height_weight.csv')
df.head()
```

```python
fig = plt.figure(figsize=(15,8))
ax = fig.gca()

ax.scatter(x=df['Height'], y=df['Weight'], c='r', s=100)
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Weight (kg)')
ax.set_title('Demographic data')
plt.show()
```

```python
color_by_sex = df['Sex'].apply(lambda x: 'r' if x == 'F' else 'b')
fig = plt.figure(figsize=(15,8))
ax = fig.gca()

ax.scatter(x=df['Height'], y=df['Weight'], c=color_by_sex, s=100)
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Weight (kg)')
ax.set_title('Demographic data')
plt.show()
```

## Historgram

<p style="line-height:1.75;font-size:16px">
A histogram is a graphical representation of the distribution of numerical data. Similar in appearance to a bar graph, the histogram condenses a data series into an easily interpreted visual by taking many data points and grouping them into logical ranges or bins. The x-axis respresents the range of the data which is split into bins that group data points in a given range. The y-axis represents the number of data points within a bin (also referred to as frequency).

```python
fig = plt.figure(figsize=(15,8))

ax = fig.gca()
df = pd.read_csv('./data/height_weight.csv')
ax.hist(df['Weight'], bins=20, color='blue')
ax.set_title('Weight Historgram')
ax.set_xlabel('Weight (kg)')
ax.set_ylabel('Count')
plt.show()
```

<p style="line-height:1.75;font-size:16px">
Since we have to distinct sub-populations in our dataset (males and females) it might be useful to look seperate them into two historgrams.

```python
df = pd.read_csv('./data/height_weight.csv')

fig = plt.figure(figsize=(15,8))

ax = fig.gca()
ax.hist(df[df['Sex'] == 'M']['Weight'], bins=20, color='blue', alpha=0.5)
ax.hist(df[df['Sex'] == 'F']['Weight'], bins=20, color='red', alpha=0.5)
ax.set_title('Weight Historgram')
ax.set_xlabel('Weight (kg)')
ax.set_ylabel('Count')
plt.show()
```

<p style="line-height:1.75;font-size:16px">
Similarly to the plot we saw before, we can also display histograms as subplots.

```python
df = pd.read_csv('./data/height_weight.csv')

fig = plt.figure(figsize=(15,8))

ax1 = fig.add_subplot(2, 1, 1)
ax1.hist(df[df['Sex'] == 'M']['Weight'], bins=20, color='blue', range=(40,100))
ax1.set_title('Weight Historgram - Males')
ax1.set_xlabel('Weight (kg)')
ax1.set_ylabel('Count')

ax2 = fig.add_subplot(2, 1, 2)
ax2.hist(df[df['Sex'] == 'F']['Weight'], bins=20, color='red', range=(40,100))
ax2.set_title('Weight Historgram - Females')
ax2.set_xlabel('Weight (kg)')
ax2.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

<p style="line-height:1.75;font-size:16px">
All of the histograms we've seen so far displayed the number of observations in each bin however this can sometimes be misleading - why? <br> It is common practice to use percentages for historgrams. This is closely related the probability density function (pdf) - an important concept in probability theory. Matplotlib offers a `density` parameter which allows us to normalize the histogram.

```python
df = pd.read_csv('./data/height_weight.csv')

fig = plt.figure(figsize=(15,8))

ax1 = fig.add_subplot(2, 1, 1)
ax1.hist(df[df['Sex'] == 'M']['Weight'], bins=20, color='blue', range=(40,100))
ax1.set_title('Weight Historgram - Males')
ax1.set_xlabel('Weight (kg)')
ax1.set_ylabel('Count')

ax2 = fig.add_subplot(2, 1, 2)
ax2.hist(df[df['Sex'] == 'M']['Weight'], bins=20, color='red', range=(40,100), density=True)
ax2.set_title('Weight Historgram - Males (Normalized)')
ax2.set_xlabel('Weight (kg)')
ax2.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

<p style="line-height:1.75;font-size:16px">
The cumulative distribution function (CDF) is another important concept in probability theory. It gives us the area under the PDF up to a certain $x$ or in other words - what is the probability that a random variable $X$ will take a value lower than or equal to $x$. We can plot the CDF by simply passing the `cumulative` parameter to the histogram method.

```python
df = pd.read_csv('./data/height_weight.csv')

fig = plt.figure(figsize=(15,8))

ax1 = fig.add_subplot(2, 1, 1)
ax1.hist(df[df['Sex'] == 'M']['Weight'], bins=20, color='blue', range=(40,100))
ax1.set_title('Weight Historgram - Males')
ax1.set_xlabel('Weight (kg)')
ax1.set_ylabel('Count')

ax2 = fig.add_subplot(2, 1, 2)
ax2.hist(df[df['Sex'] == 'M']['Weight'], bins=20, color='red', range=(40,100), density=True, cumulative=True)
ax2.set_title('Cumulative Density Function - Males')
ax2.set_xlabel('Weight (kg)')
ax2.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

<p style="line-height:1.75;font-size:16px">
Up until now we've used the `hist` method to plot the histogram without actually examining its return values. Let's check it out.

```python
df = pd.read_csv('./data/height_weight.csv')

fig = plt.figure(figsize=(15,10))

ax1 = fig.add_subplot(2, 1, 1)
histogram = ax1.hist(df[df['Sex'] == 'M']['Weight'], bins=20, color='blue', range=(40,100))
print(histogram)
```

We can see that we get back 3 objects:

2. <span style="line-height:1.75;font-size:16px">An array with the bins edges.
3. <span style="line-height:1.75;font-size:16px">A container of individual artists used to create the histogram.

## Boxplot

<p style="line-height:1.75;font-size:16px">
A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile (Q1), median, third quartile (Q3), and maximum). It can tell you about your outliers and what their values are. It can also tell you if your data is symmetrical, how tightly your data is grouped, and if and how your data is skewed.<br> The interquartile range (IQR) contains 50% of the data points. The maximum and minimum bounds (also referred to as whiskers) are usually calculated as the beginning/end of the IQR +- 1.5*IQR. It is also common to draw outliers (points that are above/below the whiskers).

```python
df = pd.read_csv('./data/height_weight.csv')

fig = plt.figure(figsize=(15,8))
ax = fig.gca()
ax.boxplot(df['Weight'])
ax.set_title('Weights Boxplot')
ax.set_ylabel('Weight (kg)')
plt.show()
```

```python
df = pd.read_csv('./data/height_weight.csv')

fig = plt.figure(figsize=(15,8))
ax = fig.gca()
ax.boxplot([df[df['Sex'] == 'M']['Weight'], df[df['Sex'] == 'F']['Weight']], labels=['Male', 'Female'])
ax.set_title('Weights Boxplot')
ax.set_ylabel('Weight (kg)')
plt.show()
```

## Bar Plot

## Interactive Plots

<p style="line-height:1.75;font-size:16px">
A cool feature of jupyter notebook is `IPywidgets` which are a set of checkboxes, sliders, dropdown boxes and other widgets that allow for user interaction with your code. Although you can use these widgets with any piece of code you have, things get interesting when you use these in combination with plots.<br>
Let's say you have the following scatter plot:

```python
df = pd.read_csv('./data/iris.csv')
df = df[df['Type'] == 'setosa']
fig = plt.figure(figsize=(15,8))
ax = fig.gca()
ax.set_title('Setosa Iris')
ax.set_xlabel('Sepal Width')
ax.set_ylabel('Sepal Legth')
ax.scatter(df['Sepal Width'], df['Sepal Length'])
plt.show()
```

<p style="line-height:1.75;font-size:16px">
Now let's say we want to draw a line that best fits the data (this is called regression, more on that next lessson). We know that a simple line takes the form of $y=ax+b$ where $a$ is the slope and $b$ in the intercept. For now let's set $a=1$ and $b=2$. Let's add this line to our plot.

```python
import numpy as np

df = pd.read_csv('./data/iris.csv')
df = df[df['Type'] == 'setosa']
fig = plt.figure(figsize=(15,8))
ax = fig.gca()
ax.set_title('Setosa Iris')
ax.set_xlabel('Sepal Width')
ax.set_ylabel('Sepal Legth')
ax.scatter(df['Sepal Width'], df['Sepal Length'])

a = 1
b = 2
x_vals = np.linspace(2, 4.5, 100)
y_vals = [a*x + b for x in x_vals]
ax.plot(x_vals, y_vals, c='r')
plt.show()
```

<p style="line-height:1.75;font-size:16px">
That's pretty good but we can probably do better. We could start guessing numbers and creating new plots but that seems tedious. It would be nice if we could just change $a$ and $b$ manually and see the changes.

```python
import numpy as np
from ipywidgets import interact

df = pd.read_csv('./data/iris.csv')
df = df[df['Type'] == 'setosa']


def plot_line(a, b=10):
    fig = plt.figure(figsize=(15,8))
    ax = fig.gca()
    ax.set_title('Setosa Iris')
    ax.set_xlabel('Sepal Width')
    ax.set_ylabel('Sepal Legth')
    ax.scatter(df['Sepal Width'], df['Sepal Length'])

    x_vals = np.linspace(2, 4.5, 100)
    y_vals = [a*x + b for x in x_vals]
    ax.plot(x_vals, y_vals, c='r')
    plt.show()
    
interact(plot_line, a=(-5,5,0.5), b=(-10,10,0.5))
```
